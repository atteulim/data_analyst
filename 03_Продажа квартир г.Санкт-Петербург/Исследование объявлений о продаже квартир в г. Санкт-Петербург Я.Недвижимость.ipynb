## Предобработка данных

В таблице очень много пропусков. И очень много столбцов - 22. Создадим новый dataFrame, в который возьмем только те столбцы, которые пригодятся нам для исследования.

Количество пропусков в столбце `floors_total` в соотношении с количеством данных небольшое, с ними ничего не будем делать. 

Пропуски в `locality_name` заполним для дальнейших расчетов как "Неизвестно".

`living_area` (она же жил.площадь) наверняка имеет корреляцию с общей площадью, выясним это ниже и попробуем заполнить пропуски.

Пропуски в `kitchen_area` (Площадь кухни) тоже можно попробовать скоррелировать с общей площадью. 

`days_exposition` пока оставим без изменений, возможно квартиры еще не проданы.

`cityCenters_nearest` врядли нужно заполнять, пропусков очень много, среднее или медианное значение могут отрицательно повлиять на решение задач.

`ceiling_height` заменим медианой.   

###  Выявление зависимостей

```python
df['total_area'].corr(df['living_area'])
```

```python
df['rooms'].corr(df['living_area'])
```

Предположение подтвердилось, существует прямая связь между жилой площадью и общей, а также между жилой площадью и количеством комнат. 

```python
medians = (df.groupby(['total_area', 'rooms'])
          .agg({'living_area':'median'}).rename(columns = {'living_area':'median_living_area'}))
df = df.merge(medians, on = ['total_area', 'rooms'])
df[['total_area', 'rooms', 'living_area', 'median_living_area']][df['living_area'].isna()] 
```

```python
df.loc[df['living_area'].isna(), 'living_area'] = df.loc[df['living_area'].isna(), 'median_living_area'] 
print("Количество пропусков после замены с помощью двух переменных:", df['living_area'].isna().sum())
```

```python
for total_area in df['total_area'].unique():
    median = df.loc[df['total_area'] == total_area, 'living_area'].median()
    df.loc[(df['living_area'].isna()) & (df['total_area'] == total_area), 'living_area'] = median
print("Количество пропусков после замены с помощью одной переменной:", df['living_area'].isna().sum())    
```

```python
for rooms in df['rooms'].unique():
    median = df.loc[df['rooms'] == rooms, 'living_area'].median()
    df.loc[(df['living_area'].isna()) & (df['rooms'] == rooms), 'living_area'] = median
print("Количество пропусков после замены с помощью одной переменной:", df['living_area'].isna().sum())    
```

```python
df['total_area'].corr(df['kitchen_area'])
```

```python
df['rooms'].corr(df['kitchen_area'])
```

Корреляция между общей площадью и площадью кухни есть, но она небольшая. Посмотрим на медианное и среднее значение столбца до и после замены пропусков. 

```python
df['kitchen_area'].mean()
```

```python
df['kitchen_area'].median()
```

```python
print("Количество пропусков до замены с помощью одной переменной:", df['kitchen_area'].isna().sum())
for total_area in df['total_area'].unique():
    median = df.loc[df['total_area'] == total_area, 'kitchen_area'].median()
    df.loc[(df['kitchen_area'].isna()) & (df['total_area'] == total_area), 'kitchen_area'] = median
print("Количество пропусков после замены с помощью одной переменной:", df['kitchen_area'].isna().sum())    
```

Метод, который нам давали преподаватели из методички работает, но справляется не со всеми пропусками, поэтому попробуем оставшиеся пропуски заменить с помощью метода fillna() и переменной, в которой будет лежать среднее арифметическое соотношение между Общей площадью и Площадью кухни

```python
kitchen_area_ratio = (df['kitchen_area'] / df['total_area']).mean()
print(kitchen_area_ratio)
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] * kitchen_area_ratio)
print(df['kitchen_area'].isna().sum())
```

```python
df['kitchen_area'].mean()
```

```python
df['kitchen_area'].median()
```

Медианное и среднее значения практически не изменились, думаю, что это хороший знак. 

Высота потолков теоретически может быть связана с ценой, обычно высокие потолки характерны для ЖК хорошего качества застройки.

```python
df['ceiling_height'].corr(df['last_price'])
```

Созависимость между ценой и высотой потолков отсутствует. Придется заменить медианным значением.

```python
median_ceiling_height = df['ceiling_height'].median()
df['ceiling_height'] = df['ceiling_height'].fillna(value=median_ceiling_height)
df['ceiling_height'].isna().sum()
```

Пропуски заменены. Проверим высоту потолков на адекватность значений, т.к там в начале были замечены слишком большие значения.

```python
df['ceiling_height'].value_counts().sample(10)
```

```python
df.loc[df['ceiling_height'] > 5.0, 'ceiling_height'] = median_ceiling_height
df.loc[df['ceiling_height'] < 2.5, 'ceiling_height'] = median_ceiling_height
df['ceiling_height'].value_counts().sample(10)
```
